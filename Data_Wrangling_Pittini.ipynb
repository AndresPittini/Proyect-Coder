**Import API **

```python
import requests
import json
import pandas as pd
url='https://api.minerstat.com/v2/hardware'
df= requests.get(url)
print(df)
```

```python
print(len(df.text))
df.text
```

```python
texto=df.text
json.loads(texto)
```

```python
jsondata=json.loads(texto)
df2=pd.DataFrame.from_dict(jsondata)
df2.head()
```

```python
df2.shape
```

```python
df2.isnull().values.any()
```

```python
df2.isnull().sum()
```

```python
df2.columns
len(df2.columns)
```

```python
percent_missing = df2.isnull().sum() * 100 / len(df2)
missing_value_df = pd.DataFrame({'Columnas': df2.columns,
                                 'Porcentaje_Missing': percent_missing})
missing_value_df.sort_values(by="Porcentaje_Missing", ascending=False)
```
